```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

We can plot univariate distribution using sns.distplot(). By default, this will draw a histogram and fit a kernel density estimate (KDE).
rug draws a small vertical tick at each observation. bins is the specification of hist bins.

```python
x = np.random.randn(100)
```

```python
sns.distplot(x)
```

```python
sns.distplot(x, kde=False)
```

```python
sns.distplot(x, kde=False, rug=True)
```

```python
sns.distplot(x, kde=False, rug=True, bins=30)
```

```python
sns.distplot(x, hist=False, rug=True)
```

```python
sns.kdeplot(x, shade=True, cbar = True, bw=2)
```

```python
sns.kdeplot(x, shade=True, cbar = True, bw=2, cut=10)
```

```python
sns.kdeplot(x, shade=True, cbar = True, bw=2, cut=0)
```

```python
sns.set()
```

```python
# Now we will see how to plot bivariate distribution.
x = tips['total_bill']
y = tips['tip']
```

```python
# Now we will plot a joint plot. It displays relationship between 2 variables (bivariate) as well as 1D profiles 
# (univariate) in the margins.
sns.jointplot(x = x, y = y)
```

```python
# By using kind we can select the kind of plot to draw. Here we have selected kind = 'hex'.
sns.jointplot(x = x, y=y, kind = 'hex')
```

```python
# Here kind = 'kde'.
sns.jointplot(x = x, y = y, kind = 'kde')
```

subplots() returns the figure and axes.
sns.cubehelix_palette() produces a colormap with linearly-decreasing (or increasing) brightness. as_cmap = True returns a matplotlib colormap instead of a list of colors. The intensity of the darkest and ligtest colours in the palette can be controlled by dark and light. As reverse = True the palette will go from dark to light.
sns.kdeplot will plot a kde plot. shade = True shades in the area under the KDE curve.

```python
f, ax = plt.subplots(figsize = (6,6))
cmap = sns.cubehelix_palette(as_cmap = True, dark = 0, light = 1, reverse= True)
sns.kdeplot(x, y, cmap = cmap, n_levels=60, shade=True)
```

```python
g = sns.jointplot(x, y, kind = 'kde', color = 'm')
g.plot_joint(plt.scatter, c = 'k', s = 30, linewidth = 1, marker = '+')
g.ax_joint.collections[0].set_alpha(0)
```

Now we will draw pair plots using sns.pairplot().By default, this function will create a grid of Axes such that each numeric variable in data will by shared in the y-axis across a single row and in the x-axis across a single column. The diagonal Axes are treated differently, drawing a plot to show the univariate distribution of the data for the variable in that column.

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris)
```

```python
# map_diag() draws the diagonal elements are plotted as a kde plot. map_offdiag() draws the non-diagonal elements as a kde plot 
# with number of levels = 10.
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels = 10)
```
