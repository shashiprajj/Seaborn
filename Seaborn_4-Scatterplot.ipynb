# Seaborn Scatter Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Import dataset from GitHub Seborn Repository
titanic_df = sns.load_dataset("titanic")
titanic_df # call titanic DataFrame
```

```python
# Draw Seaborn Scatter Plot to find relationship between age and fare
sns.scatterplot(x = "age", y = "fare", data = titanic_df)
```

```python
	
# Draw Seaborn Scatter Plot to find relationship between age and fare
sns.scatterplot(x = titanic_df.age, y = titanic_df.fare)
```

```python
# Draw Seaborn Scatter Plot to find relationship between age and fare
sns.scatterplot(x = titanic_df['age'], y = titanic_df['fare'])
```

```python
# scatter plot hue parameter
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex")
```

```python
# scatter plot hue_order parameter
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex",
               hue_order= ['female', 'male'])
```

```python
# At the bottom of the graph, it is more complex to understand, so we can work on it
plt.figure(figsize=(15,8))
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex",
               hue_order= ['female', 'male'], style="who")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex",
               hue_order= ['female', 'male'], style="who", size="who")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex",
               hue_order= ['female', 'male'], style="who", size="who", sizes=(50,200))
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x = "who", y = "fare", data = titanic_df, hue = "alive", style="alive", size="who", sizes=(50, 200))
```

```python
plt.figure(figsize=(15, 8))
sns.scatterplot(x="age", y="fare", data=titanic_df)
sns.lineplot(x="age", y="fare", data=titanic_df)
sns.barplot(x="age", y="fare", data=titanic_df)
```
