```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# We can draw regression plots with the help of sns.regplot(). The plot drawn below shows the relationship between total_bill 
#and tip.
sns.set()
sns.regplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# We can draw a linear model plot using sns.lmplot().
sns.lmplot(x = 'total_bill', y= 'tip', data = tips)
```

```python
# We can draw a plot which shows the linear relationship between size and tips.
sns.lmplot(x = 'size', y = 'tip', data = tips)
```

```python
# We can draw a plot which shows the linear relationship between size and tips.
sns.lmplot(x = 'size', y = 'tip', data = tips, x_jitter = 0.05)
```

```python
# If we set x_estimator = np.mean the dots in the above plot will be replaced by the mean and a confidence line.
sns.lmplot(x = 'size', y = 'tip', data = tips, x_estimator = np.mean)
```

```python
sns.lmplot(x = 'size', y = 'tip', data = tips, x_estimator = np.median)
```

```python
# Now we will see how to draw a plot for the data which is not linearly related. To do this we will load the anscombe dataset.
data = sns.load_dataset('anscombe')
data.head()
```

```python
# This dataset contains 4 types of data and each type contains 11 values.
```

```python
data['dataset'].value_counts()
```

```python
# Now we will draw a plot for the data of type I from the dataset. scatter_kws is used to pass additional keyword arguments.
sns.lmplot(x = 'x', y = 'y', data = data.query("dataset == 'I'"), ci = None, scatter_kws={'s': 80})
```

```python
# Now we will plot the dataset type II. We can see that it is not a linear relation. In order to fit such type of dataset, 
# we can use the order parameter. If an order is greater than 1, it estimates a polynomial regression.
sns.lmplot(x = 'x', y = 'y', data = data.query("dataset == 'II'"), ci = None, scatter_kws={'s': 80})
```

```python
sns.lmplot(x = 'x', y = 'y', data = data.query("dataset == 'II'"), ci = None, scatter_kws={'s': 80}, order=2)
```

Now we will see how to handle outliers. An outlier is a data point that differs significantly from other observations. We can go and manually remove the outlier from the dataset or we can set robust = True to nullify its effect while drawing the plot.


```python
sns.lmplot(x = 'x', y = 'y', data = data.query("dataset == 'III'"), ci = None, scatter_kws={'s': 80})
```

```python
sns.lmplot(x = 'x', y = 'y', data = data.query("dataset == 'III'"), ci = None, scatter_kws={'s': 80}, order=2)
# sns.lmplot(x = 'x', y = 'y', data = data.query("dataset == 'III'"), ci = None, scatter_kws={'s': 80}, robust=True)
```
