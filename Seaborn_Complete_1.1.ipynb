Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

Statistical analysis is a process of understanding how variables in a dataset relate to each other and how those relationships depend on other variables. Visualization can be a core component of this process because, when data are visualized properly, the human visual system can see trends and patterns that indicate a relationship.

# Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Seaborn Relplots
seaborn.relplot(*, x=None, y=None, hue=None, size=None, style=None, data=None, row=None, col=None, col_wrap=None, row_order=None, col_order=None, palette=None, hue_order=None, hue_norm=None, sizes=None, size_order=None, size_norm=None, markers=None, dashes=None, style_order=None, legend='auto', kind='scatter', height=5, aspect=1, facet_kws=None, units=None, **kwargs)

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set(style = 'darkgrid')
```

```python
# relational plot
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', style = 'size', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'size', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'size', data = tips, palette = 'ch:r=-0.8, l= 0.95')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'size')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'size', sizes = (15, 200))
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue="smoker", size = 'size', sizes = (15, 200))
```

```python
from numpy.random import randn
df = pd.DataFrame(dict(time = np.arange(500), value = randn(500).cumsum()))
df.head()
```

```python
df
```

```python
sns.relplot(x = 'time', y = 'value', data = df)
```

```python
sns.relplot(x = 'time', y = 'value', kind = 'line', data = df, sort = True)
```

```python
df = pd.DataFrame(randn(500, 2).cumsum(axis = 0), columns = ['time', 'value'])
df.head()
```

```python
sns.relplot(x = 'time', y = 'value', kind = 'line', data = df, sort = True)
```

```python
sns.relplot(x = 'time', y = 'value', kind = 'line', data = df, sort = False)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'time', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'day', data = tips)
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="time", col="day", col_wrap=2)
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day", col="time", row="sex")
```
