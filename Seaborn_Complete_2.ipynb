```python
crash_df = sns.load_dataset("car_crashes")
```

```python
crash_df
```

# Distribution Plot

```python
sns.distplot(crash_df['not_distracted'])
```

```python
sns.distplot(crash_df['not_distracted'], kde= False)
```

```python
sns.distplot(crash_df['not_distracted'], bins=25, kde=False)
```

# Joint Plot

```python
sns.jointplot(x="speeding", y="alcohol", data=crash_df)
```

```python
sns.jointplot(x="speeding", y="alcohol", data=crash_df, kind="reg")
```

```python
sns.jointplot(x="speeding", y="alcohol", data=crash_df, kind="kde")
```

```python
sns.jointplot(x="speeding", y="alcohol", data=crash_df, kind="hex")
```

# KDE

```python
sns.kdeplot(crash_df['alcohol'])
```

# Pair Plot

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

```python
sns.pairplot(tips_df, hue="sex")
```

```python
sns.pairplot(tips_df, hue="sex", palette="Blues")
```

# Rug Plot

```python
sns.rugplot(tips_df['tip'])
```

# Styling

```python
sns.set_style(style="white")
sns.jointplot(data=crash_df, x="speeding", y="alcohol", kind="reg")
```

```python
sns.set_style(style="whitegrid")
sns.jointplot(data=crash_df, x="speeding", y="alcohol", kind="reg")
```

```python
sns.set_style(style="darkgrid")
sns.jointplot(data=crash_df, x="speeding", y="alcohol", kind="reg")
```

```python
sns.set_style(style="whitegrid")
sns.set_context("paper", font_scale=1.4)
sns.jointplot(data=crash_df, x="speeding", y="alcohol", kind="reg")
```

```python
sns.set_style(style="whitegrid")
sns.set_context("talk", font_scale=1.4)
sns.jointplot(data=crash_df, x="speeding", y="alcohol", kind="reg")
```

```python
sns.set_style(style="whitegrid")
sns.set_context("poster", font_scale=1.4)
sns.jointplot(data=crash_df, x="speeding", y="alcohol", kind="reg")
```

```python
sns.set_style(style="whitegrid")
sns.set_context("paper", font_scale=2)
sns.jointplot(data=crash_df, x="speeding", y="alcohol", kind="reg")
```

```python
sns.set_style(style="whitegrid")
sns.set_context("paper", font_scale=2)
sns.jointplot(data=crash_df, x="speeding", y="alcohol", kind="reg")
sns.despine(left=True)
```

```python
sns.set_style(style="whitegrid")
sns.set_context("paper", font_scale=2)
sns.jointplot(data=crash_df, x="speeding", y="alcohol", kind="reg")
sns.despine(left=True, bottom=True)
```

# Categorical Plot

# Bar Plot

```python
sns.barplot(x="sex", y="total_bill", data=tips_df)
```

```python
sns.barplot(x="sex", y="total_bill", data=tips_df, estimator=np.median)
```

```python
sns.barplot(x="sex", y="total_bill", data=tips_df, estimator=np.cov) # covariance  
```

# Count Plot

```python
sns.countplot(x="sex", data=tips_df)
```

```python
sns.countplot(x="sex", data=tips_df, dodge=False)
```

# Box Plot 

```python
sns.boxplot(x="day", y="total_bill", data=tips_df)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips_df, hue="sex")
```

# Violin Plot (combination of box and Kde plot)

```python
sns.violinplot(x="day", y="total_bill", data=tips_df)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips_df, hue="sex")
```

```python
sns.violinplot(x="day", y="total_bill", data=tips_df, hue="sex", split=True)
```

# Strip Plot

```python
sns.stripplot(x="day", y="total_bill", data=tips_df)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips_df, jitter=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips_df, jitter=True, hue="sex")
```

```python
sns.stripplot(x="day", y="total_bill", data=tips_df, jitter=True, hue="sex", dodge=True)
```

# Swarm Plot

```python
sns.swarmplot(x="day", y="total_bill", data=tips_df)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips_df)
sns.swarmplot(x="day", y="total_bill", data=tips_df, color="white")
```

# Palette

```python
sns.set_style(style="darkgrid")
sns.set_context("talk")
sns.stripplot(x="day", y="total_bill", data=tips_df, hue="sex")
```

```python
sns.set_style(style="darkgrid")
sns.set_context("talk")
sns.stripplot(x="day", y="total_bill", data=tips_df, hue="sex", palette="magma")
```

```python
sns.set_style(style="darkgrid")
sns.set_context("talk")
sns.stripplot(x="day", y="total_bill", data=tips_df, hue="sex",palette="afmhot")
```

```python
sns.set_style(style="darkgrid")
sns.set_context("talk")
sns.stripplot(x="day", y="total_bill", data=tips_df, hue="sex", palette="coolwarm")
plt.legend(loc=(1.2,0.4))
```
